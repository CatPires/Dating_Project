```python
import pingouin as pg
```

```python
def anova_test(var_num, var_bins, dframe):
    an = pg.anova(dv=var_num, between=var_bins, data=dframe,ss_type=2,detailed=False,effsize='np2').round(3)
    return print(an)
```

```python
abc = anova_test('Self_dif',['Age_nm','Gender'],data) 
type(abc)
```

```python
## Source - 
## SS - sum of squares
## DF - degree of freedom
## MS - mean of squares
## F  - F value
## p-unc  - significance
## np2
```

```python
## The p value obtained from ANOVA:
##P-value â‰¤ 0.05: The differences between some of the means are statistically significant - there is difference 
##P-value > 0.05: The differences between the means are not statistically significant
```
